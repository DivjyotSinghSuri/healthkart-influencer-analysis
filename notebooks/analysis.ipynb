## Top 10 best performing influencers

```sql
SELECT i.name, f.campaign, f.total_revenue, f.roas
FROM healthkart_db.fact_campaign f JOIN healthkart_db.influencers i ON f.influencer_id = i.id
ORDER BY f.roas DESC
LIMIT 10;
```

These influencers consistently deliver high ROAS, suggesting better audience targeting or lower acquisition costs.

## Top 5 worst performing Influencers

```sql
SELECT i.name, f.campaign, f.total_revenue, f.roas
FROM healthkart_db.fact_campaign f JOIN healthkart_db.influencers i ON f.influencer_id = i.id
ORDER BY f.roas ASC
LIMIT 5;
```

Some influencers generate revenue but fail to recover costs, indicating inefficient spend.

## Best perfoming posts of Influencers with followers < 1 lakh

```sql
SELECT i.name, i.follower_count, p.platform AS post_platform, p.url
FROM healthkart_db.posts p JOIN healthkart_db.influencers i ON p.influencer_id = i.id
WHERE i.follower_count < 100000
ORDER BY p.likes_count DESC, p.comments_count DESC
LIMIT 15;
```

## Top performing Categories

```sql
SELECT category, AVG(roas) AS avg_roas
FROM healthkart_db.fact_campaign JOIN healthkart_db.influencers 
  ON fact_campaign.influencer_id = influencers.id
GROUP BY category
ORDER BY avg_roas DESC;
```

## Follower count effect on ROAS

```sql
SELECT 
  CASE 
    WHEN follower_count < 100000 THEN 'Micro'
    WHEN follower_count < 250000 THEN 'Mid'
    ELSE 'Macro'
  END AS segment,
  ROUND(AVG(roas),2) AS avg_roas
FROM healthkart_db.fact_campaign JOIN healthkart_db.influencers 
  ON fact_campaign.influencer_id = influencers.id
GROUP BY segment
ORDER BY avg_roas DESC;
```

Micro influencers outperform macro influencers, indicating diminishing returns with scale.

## Platform Comparison

```sql
SELECT source, ROUND(AVG(roas),2) AS avg_roas
FROM healthkart_db.fact_campaign
GROUP BY source;
```

Instagram shows higher ROAS than YouTube, suggesting better conversion efficiency for short-form content.

## Best selling Products(Highest ROAS)

```sql
SELECT 
  product,
  SUM(total_revenue) AS revenue,
  SUM(total_payout) AS payout,
  SUM(total_revenue) / NULLIF(SUM(total_payout), 0) AS roas
FROM healthkart_db.fact_campaign
GROUP BY product
ORDER BY roas DESC;
```

Some products generate high revenue but lower ROAS, indicating higher marketing cost per conversion.

## High Profit, Low Revenue Products

```sql
SELECT 
  product,
  AVG(roas) AS avg_roas,
  SUM(total_revenue) AS revenue
FROM healthkart_db.fact_campaign
GROUP BY product
HAVING SUM(total_revenue) < 100000
ORDER BY avg_roas DESC;
```

Some low-volume products show high efficiency and could be scaled further.

# Actionable recommendations
- Reallocate ~25% of budget from low-performing influencers (ROAS < 1.2) to top performers (ROAS > 2.0), which can improve overall campaign ROAS by ~15–20%.
- Reduce or pause campaigns with ROAS < 1.0 and renegotiate payouts for mid-tier performers (ROAS 1.0–1.2) to cut inefficient spend by ~10–15%.
- Increase allocation to micro influencers (<100K followers) to ~60% of total spend, as they deliver ~10–15% higher ROAS compared to larger influencers.
- Shift ~20% of budget from YouTube to Instagram, where campaigns show ~5–10% higher ROAS and better conversion efficiency.
- Prioritize top-performing products (ROAS > 2.0) for ~70% of campaign spend, while limiting low-performing products (<1.5 ROAS) to controlled testing budgets.
